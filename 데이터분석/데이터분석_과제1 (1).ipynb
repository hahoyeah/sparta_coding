# 지니뮤직의 1~50위 곡을 스크래핑 해보세요.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'}
```

```python
genie = requests.get('https://www.genie.co.kr/chart/top200?ditc=D&ymd=20220809&hh=13&rtm=Y&pg=1',headers=headers)
genie.content
```

```python
genie.text
```

```python
soup = BeautifulSoup(genie.text,'html.parser')
```

```python
soup
```

- 웹스크래핑하면서 해당부분을 copy selector을 해주면 넣어줘야할 해당 요소를 가져올 수 있다.

```python
#trs = soup.select('#body-content > div.newest-list > div > table > tbody > tr:nth-child(1)')
# 모든 요소를 가져와야하기 때문에 :nth-child(1)은 빼준다.
trs = soup.select('#body-content > div.newest-list > div > table > tbody > tr')
```

- text를 붙여주면 문자에 해당하는 부분만 가져온다.
- strip()은 띄어쓰기를 제거해주는 함수이다.

```python
for tr in trs:
  title = tr.select('td.info > a.title.ellipsis')[0].text.strip()
  rank = tr.select('td.number')[0].text[0:2].strip() #0:2로 해줘야 1 이 된다.
  artist = tr.select('td.info > a.artist.ellipsis')[0].text
  print(rank, title, artist)
```

```python
# 전체중에서 하나만 선택을 하겠다. select_one, [0]이게 사라졌다.
```

```python
df=pd.DataFrame(columns=['rank','title','artist'])
for tr in trs:
    title = tr.select_one('td.info > a.title.ellipsis').text.strip()
    rank = tr.select_one('td.number').text[0:2].strip()
    artist = tr.select_one('td.info > a.artist.ellipsis').text
    print(rank, title, artist)

    df= df.append(pd.DataFrame([[rank, title, artist]],columns=['rank','title','artist']))
```

```python
df.set_index('rank',inplace=True)
```
