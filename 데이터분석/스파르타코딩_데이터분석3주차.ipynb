## 과제

```python
# 과제에 필요한 패키지는 아래의 두 가지가 전부가 아닙니다.
# 여러분들의 필요에 따라서 패키지를 지속 추가하셔도 됩니다.
import pandas as pd
from sklearn import datasets

# 붓꽃 데이터 로드
iris_data = datasets.load_iris()
```

```python
df_data = pd.DataFrame(iris_data['data'], columns=iris_data['feature_names'])
# 랜덤으로 5개의 샘플 출력
df_data.sample(5)
```

```python
df_target = pd.DataFrame(iris_data['target'], columns=['species'])
df_target.sample(5)
```

```python
# 값의 종류를 전부 출력
df_target['species'].unique()
```

```python
df = pd.concat([df_data, df_target],axis=1) #열에 이어붙이기
df
```

```python
df.info()
```

```python
#산점도
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(df)
plt.show()
```

```python
df.pivot_table(index=['species'],aggfunc='mean')
```

```python
df.groupby('species')[['sepal length (cm)','sepal width (cm)','petal length (cm)',	'petal width (cm)']].mean()
```

```python
df_corr=df.corr(method='pearson')
df_corr
```

```python
df_corr.values
```

```python
#상관관계

# 레이블의 폰트 사이즈를 조정
sns.set(font_scale=1.0)

test_heatmap = sns.heatmap(df_corr.values, # 데이터
                           cbar = True, # 오른쪽 컬러 막대 출력 여부
                           annot = True, # 차트에 숫자를 보여줄 것인지 여부
                           annot_kws={'size' : 15}, # 숫자 출력 시 숫자 크기 조절
                           fmt = '.2f', # 숫자의 출력 소수점자리 개수 조절
                           square = True, # 차트를 정사각형으로 할 것인지
                           yticklabels= df.columns, # x축에 컬럼명 출력
                           xticklabels=df.columns) # y축에 컬럼명 출력
plt.tight_layout() #그래프 간격 유지 설명
plt.show() #그래프 표시
```
